<h1 style="color:red"> Pandas Cheat Sheet</h1>
<h2>Missing Data</h2>
<p>
    When dealing with missing data we have 3 posibilities:
    <ol>
        <li>Replacing it with the Average if it is a numeric field</li>
        <li>Use the most repetitive value if is a Categorical Field</li>
        <li>Drop the row for that specific analisis</li>
    </ol>
</p>
<h2>Exploratory Data Analysis (EDA)</h2>
<p>
    We do an EDA to:
    <ol>
        <li>summarize main characteristics of the data</li>
        <li>gain better understanding of the data set</li>
        <li>uncover relationships between different variables</li>
        <li>extract important variables for the problem we're trying to solve</li>
    </ol>
    What are the characteristics that have the most importance in the data
    <ul>
    <li>Descriptive Statistics: Defines exactly what is happening in the data or the history that the data is trying to tell.</li>
    <li>Group By: Basic of grouping data.</li>
    <li>ANOVA: Analysis of Variance.</li>
    <li>Correlation Between Variables: How a variable interacts of afects another variable</li>
    </ul>
</p>
<h2>Descriptive Statistics</h2>
<ul>
    <li>Describe basic features of data</li>
    <li>Giving short summaries about the sample and measures of the data</li>
</ul>
We can do it in Python using describe() method from pandas
To summarize the categorical data is by using the <b>value_counts()</b> method, basically is a group by the field and count the amount of rows.
<br></br><b>Box Plots</b> are a great way to visualize numeric data, since you can visualize the various distributions of the data.


```python
#dfRaw.head()
dw = dfRaw['drive-wheels'].value_counts()
dfdw = pd.DataFrame(dw)
dfdw.rename(columns={'drive-wheels':'value_counts'}, inplace=True)
dfdw.index.name='drive-wheels'
#dfdw[dfdw.index=='rwd']
dfdw.head()
```

<h2>Grouping Data</h2>
<p>
    We tend to use dataframe<b>.groupby()</b> method to divide data into categories depending on the number of variables    
</p>

```python
dftest = dfRaw.loc[:,['drive-wheels','body-style','price']]
dftest['NumericPrice'] = pd.to_numeric(dftest['price'].replace('?',''))
df_grp = dftest.groupby(['drive-wheels','body-style'], as_index=False).mean()
df_grp.head()
```

<h3>How to create Pivot Table</h3>
<p>
    DataFrame<b>.pivot(index='Yaxis, columns='Xaxis')</b><br>
    One variable displayed along the columns and the other variable displayed along the rows.
    It does not aggregate just display the information in a better way.
    
</p>

```python
df_pivot = df_grp.pivot(index='drive-wheels', columns='body-style')
df_pivot
```

<h3>Heatmap</h3>
<p>
    Plot target variable over multiple variables. <br>
    Use the matplotlib library
</p>

```python
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

ax = sns.heatmap(df_pivot,cmap='RdBu')
plt.show()
```
